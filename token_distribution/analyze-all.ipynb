# Inequality metrics of ERC20 Tokens

Data for this analysis was provided by [Ethplorer.io](https://ethplorer.io/)

```r
library(ggplot2)
library(ineq)

lorenz_curve_data <- read.csv("../data/lorenz.csv", sep = ",")
gini_data <- read.csv("../data/ineq_measures.csv", sep = ",")
all_data <- read.csv("../data/merged.csv", sep = ",")
```

## Computing the Lorenz curve for each Token

```r
ggplot(lorenz_curve_data, aes(x = p, y = L, group = symbol, color = symbol)) + 
  geom_line() +
  geom_abline(slope = 1, intercept = 0, color = "red") +
  theme_bw() +
  theme(legend.position="bottom")
```

```r
ggplot(lorenz_curve_data, aes(x = p, y = L, group = symbol, color = symbol)) + 
  geom_line() +
  geom_abline(slope = 1, intercept = 0, color = "red") +
  theme_bw() +
  facet_wrap(~symbol, nrow = 10, ncol = 5) +
  theme(legend.position="none")
```

## Computing inequality metrics

```r
ggplot(gini_data, aes (x = symbol, y = gini, color = symbol)) + 
  geom_point() +
  theme_bw() +
  theme(axis.text.x = element_text(angle = 90, hjust = 1)) +
  theme(legend.position="none")
```

```r
ggplot(gini_data, aes (x = symbol, y = theil, color = symbol)) + 
  geom_point() +
  theme_bw() +
  theme(axis.text.x = element_text(angle = 90, hjust = 1)) +
  theme(legend.position="none")
```

Theil was computed with $p = 0$

```r
ggplot(gini_data, aes (x = symbol, y = entropy, color = symbol)) + 
  geom_point() +
  theme_bw() +
  theme(axis.text.x = element_text(angle = 90, hjust = 1)) +
  theme(legend.position="none")
```

Entropy was computed with $p = 0.5$

```r
ggplot(gini_data, aes (x = symbol, y = varcoeff, color = symbol)) + 
  geom_point() +
  theme_bw() +
  theme(axis.text.x = element_text(angle = 90, hjust = 1)) +
  theme(legend.position="none")
```

```r
ggplot(gini_data, aes (x = symbol, y = atkinson, color = symbol)) + 
  geom_point() +
  theme_bw() +
  theme(axis.text.x = element_text(angle = 90, hjust = 1)) +
  theme(legend.position="none")
```

Atkinson was computed with $\epsilon = 0.5$
